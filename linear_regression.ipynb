# Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create data set: y = 5x + 5 + epsillon that is normally distributed with mean 0
dimension = 100
data = np.zeros(shape=(dimension, 2))
x = 100*np.random.random(dimension)
y = [2*i + 5 + np.random.randn(1)*10 for i in x]
data[:,0] = x
data[:,1] = y
```

```python
def linear_regression(data, data_label = 'data'):
    x_mean = np.sum(data, axis=0)[0] / data.shape[0]
    y_mean = np.sum(data, axis=0)[1] / data.shape[0]

    beta_1 = np.sum([((x[i] - x_mean)*(y[i] - y_mean)) for i in range(data.shape[0])]) / np.sum([((x[i] - x_mean)**2) for i in range(data.shape[0])])
    beta_0 = y_mean - beta_1*x_mean 
    
    y_lin_reg = [beta_1*x[i] + beta_0 for i in range(data.shape[0])]
    fig = plt.figure(figsize=(10,10))

    plt.scatter(x,y, label=data_label)
    plt.plot(x,y_lin_reg, 'g', label=f'linear regression: beta_1={beta_1}, beta_0={beta_0}')
    plt.legend()
    plt.show()
    
    print(f'y = {beta_1}x + {beta_0}')
    return 
```

```python
linear_regression(data, data_label='Data: y = 2x + 5 + epsillon')
```
